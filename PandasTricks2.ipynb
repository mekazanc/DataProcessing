Load A New Data

```
!wget -O survey_results_public.csv https://www.dropbox.com/s/w6tjen1cw4nqjd0/survey_results_public.csv?dl=0
```

```
data = pd.read_csv("survey_results_public.csv",error_bad_lines=True)
data.head()
```

```
#create grouped data by country
cdata = data.groupby(['Country'])
```

```
# get one of the groups
cdata.get_group('United Kingdom').head()
```

```
# Get Turkey's "MajorUndergrad" counts 
# cdata['MajorUndergrad'].value_counts(normalize=True) 
cdata['MajorUndergrad'].value_counts(normalize=True)['Turkey']
```

```
cdata['Respondent'].median().loc['Turkey']
```

```
cdata['Respondent'].agg(['median', 'mean']).loc['Turkey']
```

```
# cdata['StackOverflowDevices'].apply(lambda x: x.str.contains('iOS').sum())
cdata['Professional'].apply(lambda x: x.str.contains('Professional').sum()*100 / x.count())
```

```
# create number of respondents for each group. 
country_respondents = data['Country'].value_counts()
country_respondents
```

```
# create number of respondents for each group. 
country_are_prof = cdata['Professional'].apply(lambda x: x.str.contains('Professional').sum())
country_are_prof
```

```
# concat two dataframe 
newdata = pd.concat([country_respondents, country_are_prof], axis=1, sort=False)
newdata.columns = ['NoOfRespondents', 'NoOfProfs']
newdata.head()
```

```
newdata['PctOfProfs'] = (newdata['NoOfProfs']/newdata['NoOfRespondents']) * 100
newdata.head()
```

```
newdata.sort_values(by='PctOfProfs', ascending=False, inplace=True)
newdata.loc['Turkey']
```
