Load A New Data

```
data = pd.read_csv("survey_results_public.csv")
data.head()
```

```
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA','A','A'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing','B','B'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA','C','C'],
    'age': ['33', '55', '63', '36', None, None, 'Missing',np.nan,None]
}
```

```
df = pd.DataFrame(people)
df
```

```
#Note that dropna will remove "None" and "NaN" typpe elements from the dataframe. 
# axis = index will remove rows and how = any needs only 1 null element to activate rem. operation.
df.dropna(axis = 'index', how = 'any')
# axis = columns will remove columns how = all needs all elements to be null to activate rem. operation. 
```

```
# If you want to remove rows based on specific column value; 
df.dropna(axis = 'index', how = 'any', subset = ['email']) #only email column will be checked
# or 
# If you want to remove rows based on specific multiple columns ; 
df.dropna(axis = 'index', how = 'all', subset = ['email', 'first']) #only email and first column will be checked
```

It is important to mention that NaN is float type data. 

```
# This will replace NA and Missing values with np.nan elements. 
df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)

# This will replace np.nan and None elements by a given specific name. 
# df.fillna("Missing") 
# or 
# df.fillna(0)

# df['email'].replace([None], np.nan, inplace=True) #this will replace None element with np.nan.
```
