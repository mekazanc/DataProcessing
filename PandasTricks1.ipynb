Access Specific Rows and Columns

  Index Update Operations 

```
# to change index column, you can use below command. 
train.set_index('MSZoning', inplace=True)

# to reset index column you can use the following command
train.reset_index(inplace=True)

#to sort based on index
train.sort_index(ascending=True,inplace = True)
```

Filter Dataframe

```
# filter dataframe based on column condition
filtered = (train['LotShape'] == 'IR1')
filtered
```

```
# display filtered dataframe 
train[filtered].head()
```

```
# display filtered dataframe with selected columns
train.loc[filtered, ['OverallQual','OverallCond']].head()
```

```
# filter dataframe based on multiple column conditions
filtered2 = (train['LotShape'] == 'IR1') & (train['Neighborhood'] == 'Mitchel') # Note that you can replace AND(&) operator with OR(|) operator. 
train[filtered2].head()
```

```
#If you want to filter opposite conditions, you can use a "tilde" sign to create converse operation.  
train[~filtered2].head()
```

```
# filter dataframe based on s vbolumn conditions
highQual = (train['OverallQual'] > 6)  # Note that you can replace AND(&) operator with OR(|) operator. 
train[highQual].head()
```

```
train.loc[highQual, ['LotShape', 'LotConfig', 'Neighborhood', 'SaleCondition', 'OverallQual']]
```

```
# Note that filtering is also possbile by using list of categories. 
listOfCategories = ['NAmes', 'CollgCr' , 'OldTown', 'Edwards', 'Somerst']

# This is an alternative filtering operation. 
filt3 = train['Neighborhood'].isin(listOfCategories)
train.loc[filt3, 'Neighborhood']

```

```
#The final way to filter a dataframe can be done with str.contains method. 
filt4 = train['LotConfig'].str.contains('Inside') #this will generate also return NA instances. To prevent it, add one more parameter na=False
filt4 = train['LotConfig'].str.contains('Inside', na = False)

train.loc[filt4].head()
```

Changing Column Name

```
# Firstly, lets get the first 4 columns. 
train2 = train.iloc[1:10, 0:4]
train2.head()
```

```
# Change the column name. 
train2.columns = ['MSZoningNew',	'IdNew',	'MSSubClassNew',	'LotFrontageNew']
train2.head()

```

```
# Perform operation by using the for loop. 
train2.columns = [i.upper() for i in train2.columns ]
train2.head()
```

```
# It is possible to perform string operation here. 
train2.columns = train2.columns.str.replace(" ", "_") #this will not replace anything in our example. 
train2.columns = train2.columns.str.replace("NEW", "_") #this will replace "NEW" with "_" character. 


```

```
#To rename columns, you can call rename method with dictionary format.
train2.rename(columns = {"MSZONING_": "MsZoning", "MSSUBCLASS_": "MsSubClass" , "LOTFRONTAGE_": "LotFrontage"}, inplace = True)
train2.head()
```

Updating Rows

```
# If you want to update all instances in the given row. 
train2.loc[1] = ['RR', 4, 40, 180.0]
train2.head()
```

```
# If you want to update some instances in the given row. 
train2.loc[1, ['MsZoning', 'LotFrontage']] = ['RRR', 0.0]
train2.head()
```

```
#To make lowercase of the given column
train2['MsZoning'] = train2['MsZoning'].str.lower()
train2.head()
```

Common Functions used in Pandas DataFrame 

```
train.head(3)
```

```
#this will return the length of the string in the given column.  
train['Neighborhood'].apply(len)
```

```
# let's do it with a function. 
def makeUpper(data):
  return data.upper()

train['Street'] = train['Street'].apply(makeUpper)
train.head(3)
```

```
# the same operation can be done by using lambda operator
#Note that this operation is applied to Series object. 
train['Street'] = train['Street'].apply(lambda x : x.lower())
train.head()
```

```
train2.head()
```

```
# Apply the same function to the entire dataframe so that you can receive 
train2.apply(lambda x : x.sum())
```

```
# There is one more method, applymap that performs operation to each instance in the dataframe. 
train3 = train[['MSZoning', 'Street', 'Utilities', 'LotConfig', 'Neighborhood']]
train3.applymap(len).head()
```

```
# Apply the same algorithm here. 
train3.applymap(str.upper).head()
```

```
# If you want to update specific column and specific instances. 
train3['MSZoningG'] = train3['MSZoning'].replace({'RL': 'RL_NEW', 'RR': 'RR_NEW'})
train3.head()

```

```
#to update categories in the column, you can use the following options. I used "map" here but it did not work. 
train3['Neighborhood'].replace({'CollgCr' : True, 'Crawfor' : False})
```

```
people = {
    'first': ['Corey', 'Jane', 'John'], 
    'last': ['Schafer', 'Doe', 'Doe'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com']
}

df1 = pd.DataFrame(people)
df1.head()
```

```
people = {
    'test' : ['A','B','X'],
    'first': ['Selim', 'Ahmet', 'Kemal'], 
    'last': ['Yılmaz', 'Öztürk', 'Rüzgar']
}

df2 = pd.DataFrame(people)
df2.head()
```

```
#append two dataframe 
df = df1.append(df2, ignore_index=True, sort = False)
# another way is concat
# pd.concat([df1,df2], ignore_index=True)
df
```

```
#simply dropping an instance 
df.drop(index = [4,5] )
```

```
#conditional dropping 
condition = df['last'] == 'Doe'
df.drop(index = df[condition].index).reset_index(drop=True) # drop=True will prevent 
```

Sorting Columns

```
df.sort_values(by='first', ascending=False)
```

```
# Sorting with multiple columns. If there is a duplicated instances, sort them according to the second "by" parameter(column) . 
df.sort_values(by=['last', 'first'], ascending=False) #sort by last and if there are duplicated instances, sort them based on first. 
```

```
# Sorting with multiple columns. 
df.sort_values(by=['last', 'first'], ascending=[False, True]) #sort columns with different asc,desc parameters.
```

```
#To come back to the original format which is index based sorting. 
df.sort_index()
```
